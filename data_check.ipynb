```python
import pandas as pd
```

```python
import json

# Load the JSON file
with open('data/newest_data_array.json', 'r') as file:
    data = json.load(file)

# Normalize the JSON data
df = pd.json_normalize(data, sep='_', errors='ignore')

df.head(10)
```

```python
# Remove the first two arrays from the JSON data
data_without_arrays = {key: value for key, value in data.items() if key not in ['feedstocks', 'metrics']}

# Normalize the modified JSON data
df = pd.json_normalize(data_without_arrays['emissions'], sep='_', errors='ignore')

df.head(10)
```

```python
df.set_index('id', inplace=True)
```

```python
df_transposed = df.T
df_transposed.head(44)
```

Note: the head of the dataframe shows all data through CO_Urban, the tail of the dataframe, shown below, shows all of the remaining data, with the CO_Urban units included at the first row as a reference to show no missing rows.
