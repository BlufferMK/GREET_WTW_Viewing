```python
# Function to jsonify the dataframe
def jsonify_dataframe(df):
    emissions = []
    # Get the list of index values (feedstocks)
    feedstocks = df.index.tolist()

    for i in range(2, len(df)):  # Skip the first two rows (header)
        fuel_data = {"id": feedstocks[i]}  # first column as ID
        for j in range(0, len(df.columns), 3):  # Iterate over metrics
            metric = df.iloc[0, j]  # Metric name from the first row
            fuel_data[metric] = {
                "WTP": float(df.iloc[i, j]),
                "PTW": float(df.iloc[i, j + 1]),
                "WTW": float(df.iloc[i, j + 2]),
                "units": df.iloc[0, j+1],
            }
        emissions.append(fuel_data)
    return {"emissions": emissions}

# Convert the dataframe to JSON
json_data = jsonify_dataframe(df_jet_transposed)

json_data
```

```python
# Replace the text in the JSON data
json_str = json.dumps(json_data)
json_str = json_str.replace("[gCO\\u2082e/MJ]", "[gCO2e/MJ]")
json_str = json_str.replace("GHG", "CO2e")
json_data = json.loads(json_str)

json_data
```

```python
metrics = []
for emission in json_data["emissions"]:
    for key in emission.keys():
        if key != "id":  # Exclude the "id" key
            if key not in metrics:
                metrics.append(key)

metrics
```

```python
# Add the "metrics" key at the beginning of the JSON
json_data = {"metrics": metrics, **json_data}

json_data
```
