```python
import pandas as pd
```

```python
wtw_df = pd.read_csv("data\\wtw_data.csv")
wtw_df.head()  # Display the first few rows of the DataFrame
```

```python
# Group the dataframe by columns, including Feedstock and Stage as descriptors
jsonified_grouped_data = {
    column: [
        {"Feedstock": row["Feedstock"], "Stage": row["Stage"], "Value": row[column]}
        for _, row in wtw_df.iterrows()
    ]
    for column in wtw_df.columns if column not in ["Feedstock", "Stage"]
}

jsonified_grouped_data
```

```python
output_filepath = Path("data\\new_emissions_data.json")
with open(output_filepath, "w") as outfile:
    json.dump(jsonified_grouped_data, outfile, indent=4)
```
